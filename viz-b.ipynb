## Cars crashes data visulization.

```python
import seaborn as sns

df = sns.load_dataset('car_crashes')
df.head()
```

## Could be drinking alcohol while driving is related to speeding into having a crash ?

```python
sns.catplot("speeding", "alcohol", data=df, hue="not_distracted", kind="bar")
```

## Summary

## We noticed that speeding while the driver is under alcohol effect is so related that the drive would lose control and get distracted.
